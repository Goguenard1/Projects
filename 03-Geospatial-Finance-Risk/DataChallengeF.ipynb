    1) Int√©gration de l'indicateur NORIFMxAV 
</h3>

<br>

Ici, nous allons rajouter l'**indicateur de risque de feux de for√™t (<span style="color:#006400; font-weight:bold;">NORIFMxAV</span>)** √† notre **portefeuille d'actifs**.   
Pour cela, nous allons **joindre notre portefeuille d'actifs et nos donn√©es de feux de for√™t obtenues avec QGIS** en utilisant le <span style="color:#006400; font-weight:bold;">code INSEE</span> comme crit√®re de liaison.  




‚áí La <span style="color:#006400; font-weight:bold;">maturit√© de chaque pr√™t</span> d√©terminera l‚Äôhorizon temporel auquel il est expos√©, permettant ainsi d‚Äô**assigner l‚Äôindicateur de risque de feux de for√™t correspondant √† la p√©riode d‚Äôanalyse**.





‚áí Le fichier final obtenu correspondra √† notre **portefeuille d'actifs**, incluant une **nouvelle colonne** repr√©sentant l'indicateur **NORIFMxAV** s√©lectionn√© pour chaque actif.  
Ce **score** permettra d‚Äô√©valuer **l'exposition de chaque actif au risque de feux de for√™t**, en fonction de **sa localisation et de son horizon d‚Äôinvestissement**.<br><br>


```python
import pandas as pd
import numpy as np

# Charger le portefeuille d'entreprises
portefeuille = pd.read_csv('portefeuille_secteur.csv')
portefeuille.head()
```

```python
# Fonction pour charger, nettoyer et joindre les donn√©es de feux de for√™t
def ajouter_colonne_feu(portefeuille, fichier_feu, horizon):
    # Charger les donn√©es de feux de for√™t
    feu = pd.read_csv(fichier_feu, low_memory=False)
    
    # Supprimer les doublons bas√©s sur INSEE_COM
    feu = feu.drop_duplicates(subset=['INSEE_COM'])
    
    # Effectuer un left join avec le portefeuille
    resultat = pd.merge(portefeuille, feu[['INSEE_COM', 'NORIFMxAV']], 
                        how='left', left_on='Localisation (Code INSEE)', right_on='INSEE_COM')
    
    # Renommer la colonne NORIFMxAV en fonction de l'horizon
    resultat.rename(columns={'NORIFMxAV': f'NORIFMxAV_{horizon}'}, inplace=True)
    
    # Supprimer la colonne INSEE_COM ajout√©e par le merge
    resultat.drop(columns=['INSEE_COM'], inplace=True)
    
    return resultat
```

```python
feu = pd.read_csv('feux-de-forets-H1.csv', low_memory=False)
print(feu.columns)  # Check available columns
```

```python
# Ajouter les colonnes pour chaque horizon
portefeuille = ajouter_colonne_feu(portefeuille, 'feux-de-forets-H1.csv', 'H1')
portefeuille = ajouter_colonne_feu(portefeuille, 'feux-de-forets-H2.csv', 'H2')
portefeuille = ajouter_colonne_feu(portefeuille, 'feux-de-forets-H3.csv', 'H3')
```

```python
# D√©finir les conditions et les choix pour la nouvelle colonne NORIFMxAV
conditions = [
    (portefeuille['Maturit√© du pr√™t'] >= 2021) & (portefeuille['Maturit√© du pr√™t'] <= 2050),
    (portefeuille['Maturit√© du pr√™t'] >= 2051) & (portefeuille['Maturit√© du pr√™t'] <= 2070),
    (portefeuille['Maturit√© du pr√™t'] >= 2071) & (portefeuille['Maturit√© du pr√™t'] <= 2100)
]

choices = [
    portefeuille['NORIFMxAV_H1'],
    portefeuille['NORIFMxAV_H2'],
    portefeuille['NORIFMxAV_H3']
]
```

```python
# Appliquer les conditions et les choix pour cr√©er la nouvelle colonne
portefeuille['NORIFMxAV'] = np.select(conditions, choices, default=np.nan)
```

```python
# Supprimer les colonnes des horizons
portefeuille = portefeuille.drop(columns=['NORIFMxAV_H1', 'NORIFMxAV_H2', 'NORIFMxAV_H3'])
```

```python
# Sauvegarder le fichier mis √† jour
portefeuille.to_csv('portefeuille_secteur_avec_NORIFMxAV.csv', index=False)
```

Une fois ces √©tapes compl√©t√©es, nous avons **concat√©n√© les fichiers CSV obtenus** en un seul et m√™me fichier afin d‚Äôobtenir un **portefeuille contenant les 4 indicateurs DRIAS** avec lesquels nous avons travaill√©.

```python
# Charger les trois fichiers CSV
df1 = pd.read_csv('portefeuille_secteur_avec_NORCWBC.csv')
df2 = pd.read_csv('portefeuille_secteur_avec_NORIFMxAV.csv')
df3 = pd.read_csv('portefeuille_secteur_avec_NORTXHWD.csv')
df4 = pd.read_csv('portefeuille_secteur_avec_NORPAV.csv')

# Effectuer la jointure avec 'how='left', left_on='...', right_on='...' pour chaque fichier
df_combined = df1.merge(df2[['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t', 'NORIFMxAV']], 
                        how='left', 
                        left_on=['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t'], 
                        right_on=['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t'])

df_combined = df_combined.merge(df3[['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t', 'NORTXHWD']], 
                                how='left', 
                                left_on=['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t'], 
                                right_on=['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t'])

df_combined = df_combined.merge(df4[['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t', 'NORPAV']], 
                                how='left', 
                                left_on=['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t'], 
                                right_on=['Localisation (Commune)', 'Localisation (Code INSEE)', 'Maturit√© du pr√™t'])

# Sauvegarder le fichier combin√©
df_combined.to_csv('portefeuille_secteur_avec_NORCWBC_NORIFMxAV_NORTXHWD_NORPAV.csv', index=False)
```

```python
# Fonction pour attribuer les risques en fonction du secteur
def attribuer_risques(row):
    secteur = row['Secteur d\'activit√© (code NACE 2)']
    if secteur in risques_par_secteur:
        return risques_par_secteur[secteur]
    else:
        # Si le secteur n'est pas trouv√©, on attribue des risques faibles par d√©faut
        return {'S√©cheresse': 0.5, 'Inondations': 0.5, 'Feux de for√™ts': 0.5, 'Vagues de chaleur': 0.5}
```

- <span style="color:#006400; font-weight:bold;">L'encours financier</span>

- <span style="color:#006400; font-weight:bold;">L‚Äôindicateur climatique DRIAS</span>  

‚áí L‚Äôobjectif est d‚Äôobtenir **quatre scores distincts** pour chaque actif, repr√©sentant l‚Äôexposition aux **quatre al√©as climatiques**.

<br>

Pour garantir une **comparabilit√©** entre les variables, nous avons appliqu√© :  
- Une **normalisation** aux indicateurs DRIAS.  
- Une **normalisation** sur l‚Äôencours pour r√©duire l‚Äôimpact des valeurs extr√™mes et ne prendre que des valeurs normalis√©es pour ne pas biaiser notre calcul.  

<br>

La formule appliqu√©e est la suivante :

$\text{Score} = (w_{\text{encours}} \times \text{Encours}) + (w_{\text{secteur}} \times \text{Risque sectoriel}) + (w_{\text{indicateur}} \times \text{Indicateur DRIAS normalis√©})$

Avec les **pond√©rations d√©finies** :

- <span style="color:#006400; font-weight:bold;">Encours </span> ‚Üí $w_{\text{encours}} = 0.2$ 
- <span style="color:#006400; font-weight:bold;">Risque sectoriel</span> ‚Üí $w_{\text{secteur}} = 0.2$ 
- <span style="color:#006400; font-weight:bold;">Indicateur DRIAS</span> ‚Üí $w_{\text{indicateur}} = 0.5$  

<br>


```python
# Charger le fichier CSV de base
df_risques = pd.read_csv('portefeuille_secteur_avec_risques.csv')

# Fonction de normalisation
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

# Transformation logarithmique de l'encours pour r√©duire l'impact de grandes valeurs
df_risques["Encours_norm"] = normalize(df_risques["Encours (million EUR)"])

# Normalisation des indicateurs sans les ajouter au DataFrame
NORCWBC_norm = normalize(df_risques["NORCWBC"])
NORPAV_norm = normalize(df_risques["NORPAV"])
NORIFMxAV_norm = normalize(df_risques["NORIFMxAV"])
NORTXHWD_norm = normalize(df_risques["NORTXHWD"])

# D√©finition des pond√©rations pour chaque risque
w_encours = 0.2  # Poids pour l'encours
w_sector = 0.2  # Poids pour le secteur
w_indicator = 0.5  # Poids pour l'indicateur

# Calcul du score pour chaque risque en fonction de l'encours, du secteur et de l'indicateur normalis√©
df_risques["S√©cheresse"] = (
    w_encours * df_risques["Encours_norm"] +
    w_sector * df_risques["Sect S√©cheresse"] +
    w_indicator * NORCWBC_norm
)

df_risques["Inondations"] = (
    w_encours * df_risques["Encours_norm"] +
    w_sector * df_risques["Sect Inondations"] +
    w_indicator * NORPAV_norm
)

df_risques["Feux de for√™ts"] = (
    w_encours * df_risques["Encours_norm"] +
    w_sector * df_risques["Sect Feux de for√™ts"] +
    w_indicator * NORIFMxAV_norm
)

df_risques["Vagues de chaleur"] = (
    w_encours * df_risques["Encours_norm"] +
    w_sector * df_risques["Sect Vagues de chaleur"] +
    w_indicator * NORTXHWD_norm
)

# Arrondir les scores √† deux chiffres apr√®s la virgule
df_risques["S√©cheresse"] = df_risques["S√©cheresse"].round(2)*10
df_risques["Inondations"] = df_risques["Inondations"].round(2)*10
df_risques["Feux de for√™ts"] = df_risques["Feux de for√™ts"].round(2)*10
df_risques["Vagues de chaleur"] = df_risques["Vagues de chaleur"].round(2)*10

# Enregistrer le DataFrame avec les r√©sultats dans un nouveau fichier CSV sans les colonnes de normalisation
df_risques.drop(columns=["Encours_norm"], inplace=True)
df_risques
```

">
    6) Calcul des scores de vuln√©rabilit√© climatique avec composante stochastisque d'√©v√©nements extr√™mes
</h3>


Ce code mod√©lise le risque climatique en utilisant les indicateurs **NOR**, l'**exposition sectorielle**, et l'**encours financier**.  
Les risques **Feux de for√™ts** et **Inondations** sont soumis √† des **√©v√©nements extr√™mes** via un **processus de sauts (jump process)**.



Les indicateurs \( \text{NOR} \) et l'encours sont normalis√©s sur l'intervalle **\([0,2]\)** en utilisant la transformation MinMax :  

$$
X_{\text{norm}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
$$




Le **risque de base** pour chaque indicateur est calcul√© comme suit :

$$
R = \theta \cdot \text{NOR} + \beta \cdot \text{Secteur} + \gamma \cdot \log(1 + \text{Encours})
$$

avec :  
- $\theta = 0.5 $: pond√©ration de l‚Äôindicateur de risque  
- $\beta = 0.2$ : pond√©ration de l‚Äôexposition sectorielle  
- $\gamma = 0.2$  : pond√©ration de l‚Äôencours financier  

Le risque de base est appliqu√© **sans sauts** pour :  
- **S√©cheresse**  
- **Vagues de chaleur**  



Les risques **Feux de for√™ts** et **Inondations** sont modifi√©s par un **processus de Poisson** avec une **distribution log-normale** pour les amplitudes de saut.

Le **taux d'occurrence des √©v√©nements extr√™mes** d√©pend de la **maturit√© du pr√™t** :

$$
\lambda = \lambda_0 \cdot (1 + \alpha \cdot T)
$$

avec :  
- $\lambda_0 = 0.02 $ (intensit√© de base)  
- $\alpha = 0.05 $ (facteur de maturit√©)  
- T = maturit√© du pr√™t  

Les √©v√©nements sont g√©n√©r√©s selon une **loi de Poisson** :

$$
N \sim \text{Poisson}(\lambda)
$$

o√π **\( N \)** repr√©sente le **nombre d‚Äô√©v√©nements extr√™mes**.

L'amplitude des sauts suit une **loi log-normale** :

$$
J \sim \log \mathcal{N}(\mu, \sigma^2)
$$

avec :  
- $ \mu = -2$ : moyenne du saut  
- $ \sigma = 0.5 $ : √©cart-type du saut  

Un **√©v√©nement extr√™me** est attribu√© **al√©atoirement** √† **Feux de for√™ts** ou **Inondations**.



Apr√®s calcul, les scores sont **remis √† l‚Äô√©chelle [0,10]** :

$$
R_{\text{final}} = \min(10, \max(0, R \times 10))
$$

Les scores sont **arrondis √† 2 d√©cimales** pour lisibilit√©.





```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

portefeuille= pd.read_csv('portefeuille_secteur_avec_risques.csv')
# Define risk indicators and exposure factors
risk_columns = ["S√©cheresse", "Feux de for√™ts", "Inondations", "Vagues de chaleur"]
nor_columns = ["NORCWBC", "NORIFMxAV", "NORPAV", "NORTXHWD"]
sector_columns = ["Sect S√©cheresse", "Sect Feux de for√™ts", "Sect Inondations", "Sect Vagues de chaleur"]
encours_column = "Encours (million EUR)"
maturity_column = "Maturit√© du pr√™t"

portefeuille[maturity_column] = portefeuille[maturity_column] - 2020

# Normalize NOR indicators to range [0,2]
scaler = MinMaxScaler()
portefeuille[nor_columns] = scaler.fit_transform(portefeuille[nor_columns])
portefeuille[encours_column] = scaler.fit_transform(portefeuille[encours_column].values.reshape(-1, 1))

# Define simulation parameters
beta, gamma, theta = 0.2, 0.2, 0.5
lambda_base, maturity_factor = 0.02, 0.05
jump_mean, jump_stddev = -2, 0.5

# Function to compute risk, with optional jumps for "Feux de for√™ts" & "Inondations"
def compute_risk(nor_value, sector_exposure, encours_log, time_to_maturity, jump_enabled=False):
    """Compute base risk, and if jump_enabled, apply a random jump to either 'Feux de for√™ts' or 'Inondations'."""
    np.random.seed(None)
    risk = theta * nor_value + beta * sector_exposure + gamma * encours_log
    
    if jump_enabled:
        jump_intensity = lambda_base * (1 + maturity_factor * time_to_maturity)
        num_jumps = np.random.poisson(jump_intensity)
        jump_value = np.random.lognormal(jump_mean, jump_stddev) if num_jumps > 0 else 0
        impacted_risk = np.random.choice(["Feux de for√™ts", "Inondations"])
        return risk, num_jumps, jump_value, impacted_risk
    
    return risk, 0, 0, None

# Apply base risk for "S√©cheresse" & "Vagues de chaleur"
for risk_col, nor_col, sect_col in zip(["S√©cheresse", "Vagues de chaleur"], ["NORCWBC", "NORTXHWD"], ["Sect S√©cheresse", "Sect Vagues de chaleur"]):
    portefeuille[risk_col], _, _, _ = zip(*portefeuille.apply(lambda row: compute_risk(row[nor_col], row[sect_col], row[encours_column], row[maturity_column]), axis=1))

# Apply risk + jumps for "Feux de for√™ts" & "Inondations"
jump_results = portefeuille.apply(
    lambda row: compute_risk(row["NORIFMxAV"], row["Sect Feux de for√™ts"], row[encours_column], row[maturity_column], jump_enabled=True), axis=1
)

# Unpack results and apply jumps
portefeuille["Feux de for√™ts"], num_jumps, jump_values, impacted_risks = zip(*jump_results)
portefeuille["Inondations"] = portefeuille["Feux de for√™ts"]  # Copy base risk

# Apply jumps to the chosen column
portefeuille["Feux de for√™ts"] += np.where(pd.Series(impacted_risks) == "Feux de for√™ts", jump_values, 0)
portefeuille["Inondations"] += np.where(pd.Series(impacted_risks) == "Inondations", jump_values, 0)

# Store catastrophe and impacted risk columns
portefeuille["Catastrophe Naturelle"] = num_jumps

# Clip risk scores between 0 and 2
portefeuille[risk_columns] = portefeuille[risk_columns].clip(lower=0, upper=1)
portefeuille[risk_columns] = portefeuille[risk_columns]*10
portefeuille[risk_columns] = portefeuille[risk_columns].round(2)

portefeuille.drop(columns=["Encours (million EUR)"], inplace=True)
```
