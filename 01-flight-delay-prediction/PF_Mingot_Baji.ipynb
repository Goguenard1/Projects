```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import OneHotEncoder, StringIndexer
from pyspark.ml import Pipeline
from pyspark.sql import DataFrame as SparkDataFrame
from pyspark.ml.feature import VectorAssembler
from pyspark.sql.functions import when
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.sql.functions import col
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.classification import MultilayerPerceptronClassifier
from sklearn import metrics
from pyspark.ml.classification import GBTClassifier
```

```python
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Test PySpark") \
    .getOrCreate()

print(f"Spark version: {spark.version}")
```

```python
spark_df = spark.read.csv("data_fillna.csv", header=True, inferSchema=True)
```

```python
# List of columns to encode
categorical_columns = ['month', 'day_of_month', 'day_of_week', 'src_airport', 'dest_airport', 'carrier']

# Create StringIndexers and OneHotEncoders
indexers = [StringIndexer(inputCol=col, outputCol=f"{col}_index") for col in categorical_columns]
encoders = [OneHotEncoder(inputCol=f"{col}_index", outputCol=f"{col}_vec") for col in categorical_columns]

# Combine indexers and encoders into a pipeline
pipeline = Pipeline(stages=indexers + encoders)
```

```python
# Fit the pipeline and transform the data
model = pipeline.fit(spark_df)
transformed_data = model.transform(spark_df)
```

```python

feature_columns = [f"{col}_vec" for col in categorical_columns]
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")

# Transform the data to include the combined features column
final_data = assembler.transform(transformed_data)
```

<h2 style="padding:16px; color:#FFF; background:#B22222; border-radius: 10px; text-align:center">Question 18 </h2>

```python

final_data = final_data.withColumn("Delay", when(final_data["arrival_delay"] > 15, 1).otherwise(0))

final_data = final_data.drop("arrival_delay")
```

```python
final_data.select("month","day_of_month").show()
```

<h2 style="padding:16px; color:#FFF; background:#B22222; border-radius: 10px; text-align:center">Question 19 </h2>

```python
train_data, test_data = final_data.randomSplit([0.85, 0.15], seed=42)
```

```python
lr = LogisticRegression(featuresCol="features", labelCol="Delay")
lr_model = lr.fit(train_data)
predictions = lr_model.transform(test_data)

## Defined accuracy :
accuracy = BinaryClassificationEvaluator(labelCol="Delay", metricName= "accuracy")
correct_prediction = predictions.filter(col("Delay") == col("prediction")).count()
total_predictions = predictions.count()
accuracy = correct_prediction/total_predictions

# defined roc_auc :
evaluator_roc = BinaryClassificationEvaluator(labelCol="Delay", metricName="areaUnderROC")
roc_auc = evaluator_roc.evaluate(predictions)

print(f"Area Under ROC Curve (AUC): {roc_auc}")
print(f"Accuracy on Test Set: {accuracy}")
```

```python
# Tracer la courbe ROC pour l'ensemble d'entranement

trainingSummary = lr_model.summary


roc = trainingSummary.roc.toPandas()
plt.figure(figsize=(8, 6))  # Taille du graphique
plt.plot(roc['FPR'], roc['TPR'], color='blue', label="Courbe ROC", linewidth=2)
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label="Rfrence alatoire")
plt.title('Courbe ROC pour la rgression logistique', fontsize=14)
plt.xlabel('Taux de faux positifs (FPR)', fontsize=12)
plt.ylabel('Taux de vrais positifs (TPR)', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(loc='lower right')
plt.show()
```

<h2 style="padding:16px; color:#FFF; background:#B22222; border-radius: 10px; text-align:center">Question 20 </h2>

```python
# Initialiser le modle de fort alatoire
rf = RandomForestClassifier(featuresCol='features', labelCol='Delay', numTrees=10, maxDepth=5)

# Entraner le modle
rf_model = rf.fit(train_data)

# Faire des prdictions sur l'ensemble de test
predictions = rf_model.transform(test_data)

# valuer le modle de fort alatoire avec AUC
evaluator_auc = BinaryClassificationEvaluator(labelCol="Delay", metricName="areaUnderROC")
auc_test = evaluator_auc.evaluate(predictions)

## Defined accuracy :
accuracy = BinaryClassificationEvaluator(labelCol="Delay", metricName= "accuracy")
correct_prediction = predictions.filter(col("Delay") == col("prediction")).count()
total_predictions = predictions.count()
accuracy_rf = correct_prediction/total_predictions

print(f"Area Under ROC Curve (AUC): {auc_test}")
print(f"Accuracy on Test Set: {accuracy_rf}")
```

```python
# Tracer la courbe ROC
roc_df = predictions.select("Delay", "probability").toPandas()
fpr, tpr, thresholds = metrics.roc_curve(roc_df["Delay"], roc_df["probability"].apply(lambda x: x[1]))

# Cration du graphique
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label='Fort alatoire (AUC = %0.2f)' % auc_test)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('Taux de faux positifs (FPR)', fontsize=12)
plt.ylabel('Taux de vrais positifs (TPR)', fontsize=12)
plt.title('Courbe ROC - Fort Alatoire', fontsize=14)
plt.grid(True)  # Ajouter une grille
plt.legend(loc="lower right")
plt.show()
```

<h2 style="padding:16px; color:#FFF; background:#B22222; border-radius: 10px; text-align:center">Question 21 </h2>

```python
features_cols = ["month_index", "day_of_month_index", "day_of_week_index", "src_airport_index", "dest_airport_index", "carrier_index", "distance"]

# Crer un assembleur de caractristiques
assembler = VectorAssembler(
    inputCols=features_cols,  # Les colonnes spcifies
    outputCol="assembled_features"  # Colonne de caractristiques
)

# Dfinir l'architecture du rseau de neurones
layers = [7, 5, 4, 2]  # 7 entres (une pour chaque variable), 2 couches caches, 2 classes en sortie

# Crer le modle de rseau de neurones
mlp = MultilayerPerceptronClassifier(labelCol='Delay', featuresCol='assembled_features', layers=layers, blockSize=128, maxIter=100)

# Crer un pipeline avec l'assembleur et le modle
pipeline = Pipeline(stages=[assembler, mlp])

# Entraner le modle sur les donnes d'entranement
model = pipeline.fit(train_data)

# Faire des prdictions sur les donnes de test
predictions = model.transform(test_data)
```

```python
# valuation du modle - AUC (Area Under ROC)
evaluator_auc = BinaryClassificationEvaluator(labelCol='Delay', rawPredictionCol='rawPrediction', metricName='areaUnderROC')
auc = evaluator_auc.evaluate(predictions)

## Defined accuracy :
accuracy = BinaryClassificationEvaluator(labelCol="Delay", metricName= "accuracy")
correct_prediction = predictions.filter(col("Delay") == col("prediction")).count()
total_predictions = predictions.count()
accuracy_neurone = correct_prediction/total_predictions

print(f"Area Under ROC Curve (AUC): {auc}")
print(f"Accuracy on Test Set: {accuracy_neurone}")
```

```python
# Extraire les probabilits pour calculer la courbe ROC
roc_df = predictions.select("Delay", "probability").toPandas()
roc_df['probability'] = roc_df['probability'].apply(lambda x: x[1])  # Probabilit de la classe positive

# Calculer la courbe ROC
fpr, tpr, thresholds = metrics.roc_curve(roc_df['Delay'], roc_df['probability'])
roc_auc = metrics.auc(fpr, tpr)

# Tracer la courbe ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'Rseau Neuronal (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('Taux de faux positifs (FPR)', fontsize=12)
plt.ylabel('Taux de vrais positifs (TPR)', fontsize=12)
plt.title('Courbe ROC - Rseau Neuronal', fontsize=14)
plt.grid(True)
plt.legend(loc="lower right")
plt.show()
```
