```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import pandas as pd
import time

driver = webdriver.Firefox()
base_url = "https://www.mistergoodbeer.com/top-bars-sympa-paris/top-bars-pas-chers-paris-{arrondissement}-arrondissement"
data = []

for arr_num in range(1, 21):
    try:
        # Formatage de l'arrondissement
        arr_str = f"{arr_num}e" if arr_num > 1 else "1e"
        current_url = base_url.format(arrondissement=arr_str)
        
        print(f"\n{'='*40}")
        print(f"ARRONDISSEMENT {arr_str.upper()} : {current_url}")
        print(f"{'='*40}")
        
        driver.get(current_url)
        
        # Attente du conteneur principal
        WebDriverWait(driver, 25).until(
            EC.presence_of_element_located((By.XPATH, "/html/body/div/div/main/div[3]"))
        )
        
        # Scroll pour charger le contenu dynamique
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(1.5)
        
        # Détection des bars
        bars = driver.find_elements(By.XPATH, "/html/body/div/div/main/div[3]/div")
        print(f"🍻 {len(bars)} bars trouvés dans le {arr_str}")

        for i in range(1, len(bars) + 1):
            try:
                # Extraction du lien
                bar_xpath = f"/html/body/div/div/main/div[3]/div[{i}]/div/a"
                link = driver.find_element(By.XPATH, bar_xpath).get_attribute("href")
                
                if not link or "bars/" not in link:
                    continue
                
                # Nouvel onglet
                driver.switch_to.new_window('tab')
                driver.get(link)
                
                # Extraction des données
                WebDriverWait(driver, 15).until(
                    EC.presence_of_element_located((By.XPATH, "//h1"))
                )
                
                name = driver.find_element(By.XPATH, "//h1").text.split(",")[0].strip()
                
                try:
                    happy_hour = driver.find_element(
                        By.XPATH, 
                        "/html/body/div/div/main/section[1]/div[1]/div/div[2]/p/span"
                    ).text
                except NoSuchElementException:
                    happy_hour = "N/A"
                
                data.append({
                    "Arrondissement": arr_str,
                    "Nom": name,
                    "Happy Hour": happy_hour,
                    "URL": link
                })
                
                print(f"✅ {arr_str} | {name[:20]}... | {happy_hour}")
                
                # Fermeture onglet
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                
            except Exception as e:
                print(f"🚨 Erreur bar {i} : {str(e)[:50]}...")
                if len(driver.window_handles) > 1:
                    driver.close()
                    driver.switch_to.window(driver.window_handles[0])
                continue
                
    except TimeoutException:
        print(f"⏳ Timeout sur l'arrondissement {arr_str} - passage au suivant")
        continue
        
    except Exception as e:
        print(f"⚠️ Erreur majeure : {type(e).__name__} - passage au suivant")
        continue

# Nettoyage final
driver.quit()
pd.DataFrame(data).to_csv("happy_hours_paris.csv", index=False)
print(f"\n💾 Fichier généré avec {len(data)} bars sur {arr_num} arrondissements traités !")
```
